```python
import pandas as pd
from basketball_reference_web_scraper import client
import os
import glob
```

```python
ITER_YEARS = [2004, 2012]
```

```python
# Apply the transform
def apply_position_transform(r):
    pos = [str(p).split('.')[-1] for p in r['positions']]
    if len(pos) > 1:
        pos = '/'.join(pos)
    else:
        pos = pos[0]
    return pos

# Get individual player game data
for year in range(ITER_YEARS[0], ITER_YEARS[1]):
    # Create the directory
    if not os.path.exists('data/player/boxscores/%d' % year):
        os.makedirs('data/player/boxscores/%d' % year)
    # Let's do this
    players = pd.DataFrame(client.players_advanced_season_totals(season_end_year=year))
    players['positions'] = players.apply(apply_position_transform, axis=1)
    df_tot = []
    for r in players[['slug', 'team', 'name', 'positions', 'age']].values:
        slug = r[0]
        team = str(r[1]).split('.')[1]
        print("Player: %s, Team: %s (%d)" % (slug, team, year))
        player_bs = pd.DataFrame(client.regular_season_player_box_scores(
            player_identifier=slug, 
            season_end_year=year
        ))
        player_bs['slug'] = r[0]
        player_bs['name'] = r[2]
        player_bs['postions'] = r[3]
        player_bs['age'] = r[4]
        player_bs.to_csv('data/player/boxscores/%d/%s_%s_%d.csv' % (year, slug, team, year), index=False)
        df_tot.append(player_bs)
    # Concat together and output
    pd.concat(df_tot).to_csv('data/player/boxscores/totals/BR_Player_BoxScores_%d.csv' % year, index=False)
    
    # Output advanced data
    players.to_csv('data/player/advanced/BR_advanced_%d.csv' % year, index=False)
```

```python
# Date transform
def apply_date_transform(r):
    date = r['start_time']
    year, month, day = date.year, date.month, date.day
    return '%s-%s-%s' % (year, month, day)

# Iterate through and get the team box score data
for year in range(ITER_YEARS[0], ITER_YEARS[1]):
    df_sched = pd.DataFrame(client.season_schedule(season_end_year=year))
    df_sched['start_time'] = df_sched['start_time'].dt.tz_convert('US/Eastern')
    df_sched['date'] = df_sched.apply(apply_date_transform, axis=1)

    # Iterate through the days of the year
    df_tot = []
    for date in sorted(df_sched['date'].value_counts().index.values):
        print('Processing: %s' % date)
        year, month, day = date.split('-')
        # Call the client
        df_tbs = pd.DataFrame(client.team_box_scores(day=day, month=month, year=year))
        df_tbs['date'] = date
        # Merge together
        df_m = df_sched[df_sched['date'] == date]
        df_away = pd.merge(df_m, df_tbs, how='inner', left_on=['away_team', 'date'], right_on=['team', 'date'])
        df_away['location'] = 'AWAY'
        df_home = pd.merge(df_m, df_tbs, how='inner', left_on=['home_team', 'date'], right_on=['team', 'date'])
        df_home['location'] = 'HOME'
        df_ret = pd.concat([df_away, df_home])
        df_tot.append(df_ret)
    pd.concat(df_tot).to_csv('data/team/boxscores/BR_Team_BoxScores_%s.csv' % (year))
```
